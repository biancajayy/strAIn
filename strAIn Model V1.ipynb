```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End


    # Would be cool to construct 3-d angle, as opposed to 2-d
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle >180.0:
        angle = 360-angle

    return angle
```

```python
def strain_score(joint, motion_df, body_weight):
    """
    Calculate the strain score for a given joint at over time
    """

    x_col = motion_df[joint, 'x']
    y_col = motion_df[joint, 'y']
    z_col = motion_df[joint, 'z']
    angle_col = motion_df[joint, 'Angle']

    if joint in ['Left Shoulder', 'Right Shoulder', 'Left Hip', 'Right Hip']:
        secondary_angle_col = motion_df[joint, 'Secondary Angle']
        delta_secondary = np.cos(secondary_angle_col).diff().abs()
    

    delta_theta = np.cos(angle_col).diff().abs()
    delta_xyz = np.sqrt((x_col.diff().abs() ** 2) + (y_col.diff().abs() ** 2) + (z_col.diff().abs() ** 2))
    velocity = delta_xyz.diff() / 60
    acceleration = velocity.diff().abs()

    strain_df = pd.DataFrame(columns=['Frame', 'Strain Score'])

    if joint in ['Left Shoulder', 'Right Shoulder']:
        joint_angle_factor = 0.8
        joint_force_factor = 1.8
    elif joint in ['Left Hip', 'Right Hip']:
        joint_angle_factor = 1.5
        joint_force_factor = 0.8
    elif joint in ['Left Elbow', 'Right Elbow']:
        joint_angle_factor = 1
        joint_force_factor = 1
    elif joint in ['Left Knee', 'Right Knee']:
        joint_angle_factor = 2.5
        joint_force_factor = 2
    elif joint in ['Left Ankle', 'Right Ankle']:
        joint_angle_factor = 2
        joint_force_factor = 1.6



    for i in range(0, len(motion_df)-1):
        if x_col[i] is None or y_col[i] is None or angle_col[i] is None:
            strain_df.loc[i] = [i, None]
            continue
        if delta_theta[i] is None or acceleration[i] is None:
            strain_df.loc[i] = [i, None]
            continue
        strain_score = delta_theta[i] * joint_angle_factor + body_weight * acceleration[i] * joint_force_factor + np.cos(angle_col[i]) * joint_angle_factor

        if joint in ['Left Shoulder', 'Right Shoulder', 'Left Hip', 'Right Hip']:
            strain_score += delta_secondary[i] * joint_angle_factor

        strain_df.loc[i] = [i, strain_score]

   
    return strain_df
```

```python
strain_score('Left Knee', motion_df, 70).sort_values(by='Strain Score', ascending=False).head(10)
```

```python
#Finding consecutive frames with highest average strain score
joints = ['Left Shoulder', 'Right Shoulder', 'Left Elbow', 'Right Elbow', 'Left Hip', 'Right Hip', 'Left Knee', 'Right Knee', 'Left Ankle', 'Right Ankle']
joint_max_strain = {}
joint_average_strain = {}
for joint in joints:
    strain_df = strain_score(joint, motion_df, 70)
    strain_df['Rolling Strain'] = strain_df['Strain Score'].rolling(window=3).mean()
    average_strain = strain_df['Strain Score'].mean()
    max_strain = strain_df['Rolling Strain'].max()
    joint_average_strain[joint] = average_strain
    joint_max_strain[joint] = max_strain

joint_max_strain = pd.DataFrame(joint_max_strain.items(), columns=['Joint', 'Max Strain'])
joint_average_strain = pd.DataFrame(joint_average_strain.items(), columns=['Joint', 'Average Strain'])

print('Average Strain Scores:')
print(joint_average_strain)
print('\nMax Strain Scores:')
print(joint_max_strain)
```
